## Empresa Multi Segurança Eletrônica, fundada por Rafael Prudente mas atualmente gerida por sua mãe

```python
multi = despesas_all[despesas_all["CREDOR"] == "MULTI SEGURANCA ELETRONICA E PATRIMONIAL LTDA"]
multi
```

```python
sum(list(multi["LIQUIDADO"]))
```

Todos os pagamentos feitos à empresa foram durante gestões de Rafael Prudente como parlamentar da Câmara Legislativa do Distrito Federal.

```python
multi[multi["LIQUIDADO"] == multi["LIQUIDADO"].max()]
```

```python
multi_grouped = multi.groupby("Ano",as_index=False).sum(numeric_only=True)
multi_grouped
```

```python
multi_grouped["CREDOR"] = "MULTI SEGURANCA ELETRONICA E PATRIMONIAL LTDA"
```

```python
multi_grouped
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(48,40))
ax_log_multi = sns.barplot(data=multi_grouped,x="LIQUIDADO",y="CREDOR",hue="Ano",
                         palette="Spectral")
ax_log_multi.axes.set_title("Despesas do GDF com a Multi",fontsize=30) # title
ax_log_multi.set_xlabel("Valor Liquidado",fontsize=28) # x axis label
ax_log_multi.set_ylabel("Credor",fontsize=28) # y axis label
ax_log_multi.tick_params(labelsize=28)
ax_log_multi.set_xscale("log") # setting to log scale for better visualization
plt.legend(fontsize=18)
plt.show()
```

## Empresa 5 Estrelas Sistema de Segurança, administrada por Leonardo e João Vitor Prudente

```python
cinco_estrelas = despesas_all[despesas_all["CREDOR"] == "5 ESTRELAS SISTEMA DE SEGURANCA LTDA"]
cinco_estrelas
```

```python
sum(list(cinco_estrelas["LIQUIDADO"]))
```

Segundo [reportagem do Metrópoles de 2019](https://www.metropoles.com/distrito-federal/politica-df/empresas-ligadas-a-cinco-distritais-abocanharam-r-27-bilhoes-do-gdf), a empresa da família Prudente já havia recebido R$ 55.901.627,43 dos cofres públicos por meio de licitações, no período de 2007 até a data de publicação da matéria. Como os dados disponibilizados pela Administração Pública do Distrito Federal começam apenas em 2009, há dois anos faltando nesta análise, de 2007 e 2008. Mesmo assim, com estes dados, podemos ver que o valor total recebido pela empresa é muito maior, totalizando 156.375.379,12 reais.

```python
cinco_estrelas[cinco_estrelas["LIQUIDADO"] == cinco_estrelas["LIQUIDADO"].max()]
```

```python
cinco_estrelas_grouped = cinco_estrelas.groupby("Ano",as_index=False).sum(numeric_only=True)
cinco_estrelas_grouped
```

Aqui, vemos quanto dinheiro a empresa recebeu por ano. Até a data de publicação da reportagem do Metrópoles, o valor total foi de R$ 72.130.624,68. Ou seja, de janeiro de 2019 até dezembro de 2022, a empresa recebeu mais do que o dobro do valor recebido no período de 2009 a 2018.

```python
cinco_estrelas_grouped["CREDOR"] = "5 ESTRELAS SISTEMA DE SEGURANCA LTDA"
cinco_estrelas_grouped
```

```python
plt.figure(figsize=(48,40))
ax_log_5estrelas = sns.barplot(data=cinco_estrelas_grouped,x="LIQUIDADO",y="CREDOR",hue="Ano",
                               palette="Spectral")
ax_log_5estrelas.axes.set_title("Despesas do GDF com a Multi",fontsize=30) # title
ax_log_5estrelas.set_xlabel("Valor Liquidado",fontsize=28) # x axis label
ax_log_5estrelas.set_ylabel("Credor",fontsize=28) # y axis label
ax_log_5estrelas.tick_params(labelsize=28)
ax_log_5estrelas.set_xscale("log") # setting to log scale for better visualization
plt.legend(fontsize=18)
plt.show()
```

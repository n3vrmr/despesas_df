```python
def per_function(data):
    function = []
    liquidado = []
    for name in data["Função"]:
        individual_df = data[data["Função"] == name]
        valor = sum(individual_df["Liquidado"])
        individual_df = individual_df.drop_duplicates("Função")
        if name not in function:
            function.append(individual_df["Função"].values[0])
            liquidado.append(valor)
    next_df = pd.DataFrame({"Função":function,
                            "Liquidado":liquidado})
    new_df = next_df.sort_values(by="Liquidado",ascending=False,ignore_index=True)
    return new_df
```

```python
dpf_2009 = despesas_2009.groupby("Função",as_index=False).sum(numeric_only=True)
dpf_2009 = dpf_2009.sort_values(by="Liquidado",ascending=False,ignore_index=True)
dpf_2009
```

```python
def extra_columns(data,year,governor):
    years = []
    name = []
    elections = []
    for i in range(0,len(data)):
        years.append(year)
        name.append(governor)
        if year in [2010,2014,2018,2022]:
            elections.append(True)
        else:
            elections.append(False)
    data["Ano"] = years
    data["Governador"] = name
    data["Ano Eleitoral"] = elections
    return data
```
